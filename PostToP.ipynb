```python
import re
import requests
import os
from google.colab import userdata
API_KEY = userdata.get('yt_api')


def get_n_video_ids(query, depth=5):
    video_ids = set()
    page_tokens = []

    for _ in range(depth):
        last_page_token = page_tokens[-1] if page_tokens else ""
        url = f"https://youtube.googleapis.com/youtube/v3/search?part=snippet&maxResults=50&q={query}&relevanceLanguage=en&pageToken={last_page_token}&type=video&key={API_KEY}"
        headers = {'Accept': 'application/json'}

        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to fetch data: {response}")
            print(last_page_token)
            break

        video_data = response.json()
        items = video_data.get('items', [])
        new_video_ids = [item['id']['videoId'] for item in items]
        video_ids.update(new_video_ids)

        next_page_token = video_data.get('nextPageToken')
        if not next_page_token or next_page_token in page_tokens:
            break

        page_tokens.append(next_page_token)

    return list(video_ids)

links = [
    # "first time listening to music",
    "first time listening to album",
    "first time on concert",
    "analysing music",
    "learning musical instruments",
    "artists react",
    "lyrics analysis",
    "history of songs",
    "histroy of music",
    "songwriting tips",
    "vocal training",
    "dance videos",
    "life of a producer",
    "music documentary",
    "create your first track",
         ]

for link in links:
  print(link)
  big_video_ids.update(get_n_video_ids(link))

len(big_video_ids)
```

```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='auto', target='en')

def detect_and_translate(text):
    def convert_to_chunks(text,size):
      size = size-1
      chunks = []
      while len(text) > size:
          split_index = text[:size].rfind(' ')
          if split_index == -1:
              split_index = size
          chunks.append(text[:split_index])
          text = text[split_index:].strip()
      chunks.append(text)
      return chunks
    if len(text) >= 5000:
        chunks = convert_to_chunks(text,5000)
        translated_chunks = [translator.translate(chunk) for chunk in chunks]
        translated = " ".join(translated_chunks)
        return translated
    translated = translator.translate(text)
    return translated
```

```python
from tqdm.notebook import tqdm

# Initialize the progress bar
pbar = tqdm(total=len(dataset))

def process_row(row):
    lang = row['Language']
    if lang == 'en':
        pbar.update(1)
        return row

    lang = lang if lang else 'unknown'
    try:
        row['Description'] = detect_and_translate(row['Description'])
        row['Title'] = detect_and_translate(row['Title'])
        row['Language'] = 'en'
    except Exception as e:
        print(f"Error processing row {row.name}: {e}")
    finally:
        pbar.update(1)
    return row

# Apply the function row-wise
dataset = dataset.apply(process_row, axis=1)

# Close the progress bar
pbar.close()
```

### Youtube Id≈ë m√°sodpercre konvert√°l√°sa
YT szereti ilyen specifikusan t√°rolni a vide√≥ hossz√°t, viszont a modelnek numberikus adat kell, sz√≥val √°tkonvert√°ljuk m√°sodperc√©.

ex. `PT1H2M3S` -> `3723 mp`

```python
import re

def pp_convert_duration(duration):
    match = re.match(r"PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?", duration)
    if not match:
        return None

    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0
    seconds = int(match.group(3)) if match.group(3) else 0

    return hours * 3600 + minutes * 60 + seconds
```

```python
unit_test(pp_convert_duration,"PT1H2M3S", 3723)
unit_test(pp_convert_duration,"PT2M3S", 123)
```

URL-ek egyszer≈±s√≠t√©se, hogy csak a domain jelenjen meg a sz√∂vegben.

```python
from re import findall
from tldextract import extract

def get_second_level_domain(url):
    extracted = extract(url)
    return extracted.domain


def pp_replace_links_with_sld(text):
    urls = findall(r'(https?:[ |\n]?\/\/[ |\n]?[^\s]+)', text,flags=re.IGNORECASE)
    urls_starting_with_www = findall(r'www\.[^\s]+', text,flags=re.IGNORECASE)
    urls = urls + urls_starting_with_www
    urls = [(x,"".join(x.split(" "))) for x in urls]
    for intext,url in urls:
      sld = get_second_level_domain(url)
      text = text.replace(intext, sld)
    return " "+text+" "
```

### Adathalmaz augment√°l√°sa szinon√≠m√°kkal (csak adathalmazon)
Kicsi adathalmazt sz√≥ synon√≠m√°kkal augment√°lunk hogy t√∂bb tanul√≥ adata legyen a modellnek.

```python
import nltk
from nltk.corpus import wordnet
import random

nltk.download('wordnet')

def augment_text_with_synonyms(text, percentage):
    words = text.split()
    num_words_to_change = int(len(words) * percentage)
    randomly_selected_words = random.sample(words, num_words_to_change)
    augmented_words = []

    for i, word in enumerate(words):
        if word in randomly_selected_words:
              synonyms = []
              for syn in wordnet.synsets(word):
                  for l in syn.lemmas():
                      synonym = l.name().replace("_", " ").lower()
                      if synonym.isalpha():
                          synonyms.append(synonym)
              if synonyms:
                  new_word = random.choice(synonyms)
              else:
                  new_word = word
              augmented_words.append(new_word)
              continue

        augmented_words.append(word)

    return " ".join(augmented_words).lower()
```

```python
import re

def pp_remove_consecutive_duplicates(text):
  words = text.split()
  new_words = []
  for i in range(len(words)):
    if i == 0 or words[i] != words[i - 1]:
      new_words.append(words[i])

  return ' '.join(new_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def remove_similar_rows(df, threshold=0.9):
    vectorizer = TfidfVectorizer()
    title_description = df['Title'] + " " + df['Description']
    tfidf_matrix = vectorizer.fit_transform(title_description)
    similarity_matrix = cosine_similarity(tfidf_matrix)
    rows_to_remove = set()
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            if similarity_matrix[i, j] >= threshold:
                if df['Is Music'][i] != df['Is Music'][j]:
                    continue
                rows_to_remove.add(j)
    new_df = df.drop(index=list(rows_to_remove)).reset_index(drop=True)
    return new_df

dataset = dataset.reset_index(drop=True)
print(f"Eredeti adathalmaz hossza: {len(dataset)}")
dataset = remove_similar_rows(dataset)
print(f"√öj adathalmaz hossza: {len(dataset)}")
```

```python
class ITokenizer:
  name = "Tokenizer"
  def train(self, text):
    pass

  def encode(self, text):
    pass

  def encode_batch(self, texts):
    pass
```

```python
class TokenizerNone(ITokenizer):
  def encode(self, text):
    return text

  def encode_batch(self, texts):
    return texts
```

### Sz√≥ alap√∫ tokeniz√°l√°s
Sz√∂veg szavakra bont√°sa egy-egy tokenk√©nt.

**P√©lda**: `Ez egy p√©lda` ‚Üí `['Ez', 'egy', 'p√©lda']`

```python
from nltk.tokenize import word_tokenize
class TokenizerWord(ITokenizer):
  name = "TokenizerWord"
  def encode(self, text):
    return word_tokenize(text)

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

### N-Gram alap√∫ tokeniz√°l√°s
Sz√∂veg bont√°sa t√∂bb sz√≥b√≥l √°ll√≥ tokenekre.

**P√©lda**: `Ez egy p√©lda` ‚Üí `['Ez egy', 'egy p√©lda']`

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

class TokenizerNgram(ITokenizer):
  name = "TokenizerNgram"
  def __init__(self, ngram_range=(1, 2)):
    self.name = f"TokenizerNgram_{ngram_range[0]}_{ngram_range[1]}"
    self.ngram_range = ngram_range

  def ngram_tokenize_text_range(self,text):
    tokens = word_tokenize(text)
    ngrams_list = []
    for n in range(self.ngram_range[0], self.ngram_range[1] + 1):
      ngrams_list.extend([' '.join(gram) for gram in ngrams(tokens, n)])
    return ngrams_list

  def encode(self, text):
    return self.ngram_tokenize_text_range(text)

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

### BPE alap√∫ tokeniz√°l√°s
Byte Pair Encoding (BPE) alkalmaz√°sa a tokenek el≈ë√°ll√≠t√°s√°hoz, azaz a szavak szegment√°l√°sa gyakori karaktersorozatokra

**P√©lda**: `Ez egy p√©lda` ‚Üí `['Ez', 'egy', 'p√©l', 'da']`

```python
import tokenizers
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

class TokenizerBPE(ITokenizer):
  name = "TokenizerBPE"
  def __init__(self, vocab_size=10000):
    self.name = f"TokenizerBPE_{vocab_size}"
    self.vocab_size = vocab_size
    self.tokenizer = None

  def train(self, text):
    self.tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
    self.tokenizer.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=self.vocab_size, special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
    self.tokenizer.train_from_iterator(text, trainer=trainer)

  def encode(self, text):
    return self.tokenizer.encode(text).tokens

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

```python
class IVectorizer:
  name = "Vectorizer"
  def train(self, text):
    pass

  def encode(self, text):
    pass

  def encode_batch(self, texts):
    pass

  def getDict(self):
    pass
```

### Bag of Words met√≥dus
A Bag-of-Words (BoW) m√≥dszer egy sz√∂veg reprezent√°l√°s√°ra szolg√°l√≥ elj√°r√°s, amely a szavak el≈ëfordul√°s√°t sz√°molja vagy jelenl√©t√ºket jel√∂li egy adott sz√∂vegben, figyelmen k√≠v√ºl hagyva a szavak sorrendj√©t √©s jelent√©s√©t.

#### El≈ëfordul√°si gyakoris√°g
A szavak el≈ëfordul√°s√°nak gyakoris√°g√°t sz√°ml√°l√≥ vektoriz√°l√°s

**P√©lda**: `a sz√∂veg sz√∂veg szavak` ‚Üí `{'a': 1, 'sz√∂veg': 2, 'szavak': 1, 'kecske': 0}`

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

class VectorizerCount(IVectorizer):
  def __init__(self,max_features=5000):
    self.name = f"VectorizerCount_{max_features}"
    self.vectorizer = CountVectorizer(tokenizer=lambda doc: doc,preprocessor=lambda doc:doc, lowercase=False,max_features=max_features)

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text).toarray()[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts).toarray()

  def getDict(self):
    voc = self.vectorizer.vocabulary_
    voc = dict(sorted(voc.items(), key=lambda item: item[1]))
    return voc

```

#### TF-IDF
A *TF-IDF (Term Frequency-Inverse Document Frequency)* egy sz√∂veg reprezent√°ci√≥s m√≥dszer, amely a szavak relat√≠v fontoss√°g√°t m√©ri egy dokumentumban, figyelembe v√©ve, hogy azok mennyire gyakoriak az adott dokumentumban (TF) √©s mennyire ritk√°k az eg√©sz korpuszban (IDF). Ez√°ltal kiemeli az adott dokumentumra jellemz≈ë, de nem √°ltal√°nos szavakat.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

class VectorizerTFIDF(IVectorizer):
  def __init__(self,max_features=5000):
    self.name = f"VectorizerTFIDF_{max_features}"
    self.vectorizer = TfidfVectorizer(tokenizer=lambda doc: doc,preprocessor=lambda doc:doc, lowercase=False,max_features=max_features)

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text).toarray()[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts).toarray()

  def getDict(self):
    voc = self.vectorizer.vocabulary_
    voc = dict(sorted(voc.items(), key=lambda item: item[1]))
    return voc
```

#### C√≠mkealap√∫ k√≥dol√°s
Kateg√≥ri√°k jelenl√©t√©nek bin√°ris vektoriz√°l√°sa.

**P√©lda**: `['Rock']` ‚Üí `{"Pop": False, "Rock": True}`

```python
from sklearn.preprocessing import MultiLabelBinarizer

class VectorizerLabel(IVectorizer):
  def __init__(self,max_features=0):
    self.name = "VectorizerLabel"
    self.vectorizer = MultiLabelBinarizer()

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text)[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts)

  def getDict(self):
    voc = self.vectorizer.classes_
    voc = dict(zip(voc, range(len(voc))))
    return voc
```

### Embedding met√≥dus

#### Word2Vec

```python
import numpy as np
from gensim.models import Word2Vec
import multiprocessing

class VectorizerWord2Vec(IVectorizer):
  def __init__(self,max_features=300):
    self.name = f"VectorizerWord2Vec_{max_features}"
    self.max_features = max_features
    pass

  def train(self,dataset):
    max_workers = multiprocessing.cpu_count()
    self.vectorizer = Word2Vec(sentences=dataset, vector_size=self.max_features, window=3, min_count=1, workers=max_workers)

  def encode(self,text):
    doc_vec = np.zeros(self.vectorizer.vector_size)
    word_count = 0
    for word in text:
      if word in self.vectorizer.wv:
        doc_vec += self.vectorizer.wv[word]
        word_count += 1
    if word_count > 0:
      doc_vec /= word_count
    return doc_vec

  def encode_batch(self,texts):
    return np.array([self.encode(text) for text in texts])

  def getDict(self):
    return self.vectorizer.wv.key_to_index
```

```python
def encode_df(dataset,tokenizer,vectorizer):
  train,val,test = dataset

  tokenizer.train(train)
  train_token = [tokenizer.encode(x) for x in train]
  val_token = [tokenizer.encode(x) for x in val]
  if test is not None:
    test_token = [tokenizer.encode(x) for x in test]

  vectorizer.train(train_token)
  train_vectors = vectorizer.encode_batch(train_token)
  val_vectors = vectorizer.encode_batch(val_token)
  if test is not None:
    test_vectors = vectorizer.encode_batch(test_token)
  else:
    test_vectors = None

  return train_vectors,val_vectors,test_vectors
```

```python
def calculate_confidence(predicted, threshold):
    if predicted >= threshold:
        confidence = (predicted - threshold) / (1 - threshold)
    else:
        confidence = (threshold - predicted) / threshold

    return round(confidence * 100, 2)
```
