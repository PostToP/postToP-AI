```python
import re
import requests
import os
from google.colab import userdata
API_KEY = userdata.get('yt_api')


def get_n_video_ids(query, depth=5):
    video_ids = set()
    page_tokens = []

    for _ in range(depth):
        last_page_token = page_tokens[-1] if page_tokens else ""
        url = f"https://youtube.googleapis.com/youtube/v3/search?part=snippet&maxResults=50&q={query}&relevanceLanguage=en&pageToken={last_page_token}&type=video&key={API_KEY}"
        headers = {'Accept': 'application/json'}

        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to fetch data: {response}")
            print(last_page_token)
            break

        video_data = response.json()
        items = video_data.get('items', [])
        new_video_ids = [item['id']['videoId'] for item in items]
        video_ids.update(new_video_ids)

        next_page_token = video_data.get('nextPageToken')
        if not next_page_token or next_page_token in page_tokens:
            break

        page_tokens.append(next_page_token)

    return list(video_ids)

links = [
    # "first time listening to music",
    "first time listening to album",
    "first time on concert",
    "analysing music",
    "learning musical instruments",
    "artists react",
    "lyrics analysis",
    "history of songs",
    "histroy of music",
    "songwriting tips",
    "vocal training",
    "dance videos",
    "life of a producer",
    "music documentary",
    "create your first track",
         ]

for link in links:
  print(link)
  big_video_ids.update(get_n_video_ids(link))

len(big_video_ids)
```

```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='auto', target='en')

def detect_and_translate(text):
    def convert_to_chunks(text,size):
      size = size-1
      chunks = []
      while len(text) > size:
          split_index = text[:size].rfind(' ')
          if split_index == -1:
              split_index = size
          chunks.append(text[:split_index])
          text = text[split_index:].strip()
      chunks.append(text)
      return chunks
    if len(text) >= 5000:
        chunks = convert_to_chunks(text,5000)
        translated_chunks = [translator.translate(chunk) for chunk in chunks]
        translated = " ".join(translated_chunks)
        return translated
    translated = translator.translate(text)
    return translated
```

```python
from tqdm.notebook import tqdm

# Initialize the progress bar
pbar = tqdm(total=len(dataset))

def process_row(row):
    lang = row['Language']
    if lang == 'en':
        pbar.update(1)
        return row

    lang = lang if lang else 'unknown'
    try:
        row['Description'] = detect_and_translate(row['Description'])
        row['Title'] = detect_and_translate(row['Title'])
        row['Language'] = 'en'
    except Exception as e:
        print(f"Error processing row {row.name}: {e}")
    finally:
        pbar.update(1)
    return row

# Apply the function row-wise
dataset = dataset.apply(process_row, axis=1)

# Close the progress bar
pbar.close()
```

### Youtube Idő másodpercre konvertálása
YT szereti ilyen specifikusan tárolni a videó hosszát, viszont a modelnek numberikus adat kell, szóval átkonvertáljuk másodpercé.

ex. `PT1H2M3S` -> `3723 mp`

```python
import re

def pp_convert_duration(duration):
    match = re.match(r"PT(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?", duration)
    if not match:
        return None

    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0
    seconds = int(match.group(3)) if match.group(3) else 0

    return hours * 3600 + minutes * 60 + seconds
```

```python
unit_test(pp_convert_duration,"PT1H2M3S", 3723)
unit_test(pp_convert_duration,"PT2M3S", 123)
```

ex. "𝕮𝖔𝖑𝖑𝖆𝖗 𝖔𝖋 𝕿𝖗𝖚𝖙𝖍" => "Collar of Truth"

```python
from unicodedata import normalize,combining
from ftfy import fix_text
from unidecode import unidecode

def pp_remove_fancy_text(text):
    text = fix_text(text)
    normalized_text = normalize("NFKD", text)
    ascii_text = unidecode("".join([c for c in normalized_text if not combining(c)]))
    return ascii_text
```

### Linkek egyszerűsítése
URL-ek egyszerűsítése, hogy csak a domain jelenjen meg a szövegben.

**Példa**: https://www.example.com/oldal → example


```python
!pip install tldextract
```

```python
from re import findall
from tldextract import extract

def get_second_level_domain(url):
    extracted = extract(url)
    return extracted.domain


def pp_replace_links_with_sld(text):
    urls = findall(r'(https?:[ |\n]?\/\/[ |\n]?[^\s]+)', text,flags=re.IGNORECASE)
    urls_starting_with_www = findall(r'www\.[^\s]+', text,flags=re.IGNORECASE)
    urls = urls + urls_starting_with_www
    urls = [(x,"".join(x.split(" "))) for x in urls]
    for intext,url in urls:
      sld = get_second_level_domain(url)
      text = text.replace(intext, sld)
    return " "+text+" "
```

### Adathalmaz augmentálása szinonímákkal (csak adathalmazon)
Kicsi adathalmazt szó synonímákkal augmentálunk hogy több tanuló adata legyen a modellnek.

```python
import nltk
from nltk.corpus import wordnet
import random

nltk.download('wordnet')

def augment_text_with_synonyms(text, percentage):
    words = text.split()
    num_words_to_change = int(len(words) * percentage)
    randomly_selected_words = random.sample(words, num_words_to_change)
    augmented_words = []

    for i, word in enumerate(words):
        if word in randomly_selected_words:
              synonyms = []
              for syn in wordnet.synsets(word):
                  for l in syn.lemmas():
                      synonym = l.name().replace("_", " ").lower()
                      if synonym.isalpha():
                          synonyms.append(synonym)
              if synonyms:
                  new_word = random.choice(synonyms)
              else:
                  new_word = word
              augmented_words.append(new_word)
              continue

        augmented_words.append(word)

    return " ".join(augmented_words).lower()
```

### Hasonló sorok törlése (csak adathalmazon)
Ha 90%ban egyezik 2 sor és ugynaz a labelje akkor töröljük.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def remove_similar_rows(df, threshold=0.9):
    vectorizer = TfidfVectorizer()
    title_description = df['Title'] + " " + df['Description']
    tfidf_matrix = vectorizer.fit_transform(title_description)
    similarity_matrix = cosine_similarity(tfidf_matrix)
    rows_to_remove = set()
    for i in range(len(df)):
        for j in range(i + 1, len(df)):
            if similarity_matrix[i, j] >= threshold:
                if df['Is Music'][i] != df['Is Music'][j]:
                    continue
                rows_to_remove.add(j)
    new_df = df.drop(index=list(rows_to_remove)).reset_index(drop=True)
    return new_df

dataset = dataset.reset_index(drop=True)
print(f"Eredeti adathalmaz hossza: {len(dataset)}")
dataset = remove_similar_rows(dataset)
print(f"Új adathalmaz hossza: {len(dataset)}")
```

```python
class ITokenizer:
  name = "Tokenizer"
  def train(self, text):
    pass

  def encode(self, text):
    pass

  def encode_batch(self, texts):
    pass
```

```python
class TokenizerNone(ITokenizer):
  def encode(self, text):
    return text

  def encode_batch(self, texts):
    return texts
```

### Szó alapú tokenizálás
Szöveg szavakra bontása egy-egy tokenként.

**Példa**: `Ez egy példa` → `['Ez', 'egy', 'példa']`

```python
from nltk.tokenize import word_tokenize
class TokenizerWord(ITokenizer):
  name = "TokenizerWord"
  def encode(self, text):
    return word_tokenize(text)

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

### N-Gram alapú tokenizálás
Szöveg bontása több szóból álló tokenekre.

**Példa**: `Ez egy példa` → `['Ez egy', 'egy példa']`

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

class TokenizerNgram(ITokenizer):
  name = "TokenizerNgram"
  def __init__(self, ngram_range=(1, 2)):
    self.name = f"TokenizerNgram_{ngram_range[0]}_{ngram_range[1]}"
    self.ngram_range = ngram_range

  def ngram_tokenize_text_range(self,text):
    tokens = word_tokenize(text)
    ngrams_list = []
    for n in range(self.ngram_range[0], self.ngram_range[1] + 1):
      ngrams_list.extend([' '.join(gram) for gram in ngrams(tokens, n)])
    return ngrams_list

  def encode(self, text):
    return self.ngram_tokenize_text_range(text)

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

### BPE alapú tokenizálás
Byte Pair Encoding (BPE) alkalmazása a tokenek előállításához, azaz a szavak szegmentálása gyakori karaktersorozatokra

**Példa**: `Ez egy példa` → `['Ez', 'egy', 'pél', 'da']`

```python
!pip install tokenizers
```

```python
import tokenizers
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

class TokenizerBPE(ITokenizer):
  name = "TokenizerBPE"
  def __init__(self, vocab_size=10000):
    self.name = f"TokenizerBPE_{vocab_size}"
    self.vocab_size = vocab_size
    self.tokenizer = None

  def train(self, text):
    self.tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
    self.tokenizer.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=self.vocab_size, special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
    self.tokenizer.train_from_iterator(text, trainer=trainer)

  def encode(self, text):
    return self.tokenizer.encode(text).tokens

  def encode_batch(self, texts):
    return [self.encode(text) for text in texts]
```

## Vektorizálás

```python
class IVectorizer:
  name = "Vectorizer"
  def train(self, text):
    pass

  def encode(self, text):
    pass

  def encode_batch(self, texts):
    pass

  def getDict(self):
    pass
```

### Bag of Words metódus
A Bag-of-Words (BoW) módszer egy szöveg reprezentálására szolgáló eljárás, amely a szavak előfordulását számolja vagy jelenlétüket jelöli egy adott szövegben, figyelmen kívül hagyva a szavak sorrendjét és jelentését.

#### Előfordulási gyakoriság
A szavak előfordulásának gyakoriságát számláló vektorizálás

**Példa**: `a szöveg szöveg szavak` → `{'a': 1, 'szöveg': 2, 'szavak': 1, 'kecske': 0}`

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

class VectorizerCount(IVectorizer):
  def __init__(self,max_features=5000):
    self.name = f"VectorizerCount_{max_features}"
    self.vectorizer = CountVectorizer(tokenizer=lambda doc: doc,preprocessor=lambda doc:doc, lowercase=False,max_features=max_features)

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text).toarray()[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts).toarray()

  def getDict(self):
    voc = self.vectorizer.vocabulary_
    voc = dict(sorted(voc.items(), key=lambda item: item[1]))
    return voc

```

#### TF-IDF
A *TF-IDF (Term Frequency-Inverse Document Frequency)* egy szöveg reprezentációs módszer, amely a szavak relatív fontosságát méri egy dokumentumban, figyelembe véve, hogy azok mennyire gyakoriak az adott dokumentumban (TF) és mennyire ritkák az egész korpuszban (IDF). Ezáltal kiemeli az adott dokumentumra jellemző, de nem általános szavakat.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

class VectorizerTFIDF(IVectorizer):
  def __init__(self,max_features=5000):
    self.name = f"VectorizerTFIDF_{max_features}"
    self.vectorizer = TfidfVectorizer(tokenizer=lambda doc: doc,preprocessor=lambda doc:doc, lowercase=False,max_features=max_features)

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text).toarray()[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts).toarray()

  def getDict(self):
    voc = self.vectorizer.vocabulary_
    voc = dict(sorted(voc.items(), key=lambda item: item[1]))
    return voc
```

#### Címkealapú kódolás
Kategóriák jelenlétének bináris vektorizálása.

**Példa**: `['Rock']` → `{"Pop": False, "Rock": True}`

```python
from sklearn.preprocessing import MultiLabelBinarizer

class VectorizerLabel(IVectorizer):
  def __init__(self,max_features=0):
    self.name = "VectorizerLabel"
    self.vectorizer = MultiLabelBinarizer()

  def train(self, dataset):
    self.vectorizer.fit(dataset)

  def encode(self, text):
    text = [text]
    return self.vectorizer.transform(text)[0]

  def encode_batch(self, texts):
    return self.vectorizer.transform(texts)

  def getDict(self):
    voc = self.vectorizer.classes_
    voc = dict(zip(voc, range(len(voc))))
    return voc
```

### Embedding metódus

#### Word2Vec

```python
import numpy as np
from gensim.models import Word2Vec
import multiprocessing

class VectorizerWord2Vec(IVectorizer):
  def __init__(self,max_features=300):
    self.name = f"VectorizerWord2Vec_{max_features}"
    self.max_features = max_features
    pass

  def train(self,dataset):
    max_workers = multiprocessing.cpu_count()
    self.vectorizer = Word2Vec(sentences=dataset, vector_size=self.max_features, window=3, min_count=1, workers=max_workers)

  def encode(self,text):
    doc_vec = np.zeros(self.vectorizer.vector_size)
    word_count = 0
    for word in text:
      if word in self.vectorizer.wv:
        doc_vec += self.vectorizer.wv[word]
        word_count += 1
    if word_count > 0:
      doc_vec /= word_count
    return doc_vec

  def encode_batch(self,texts):
    return np.array([self.encode(text) for text in texts])

  def getDict(self):
    return self.vectorizer.wv.key_to_index
```

```python
from sentence_transformers import SentenceTransformer
class VectorizerBert(IVectorizer):
  def __init__(self):
    self.name = "VectorizerBert"
    self.vectorizer = SentenceTransformer('all-MiniLM-L6-v2')

  def train(self,dataset):
    pass

  def encode(self,text):
    if len(text) <= 1:
      return np.zeros(384)
    text = [text]
    return self.vectorizer.encode(text)[0]

  def encode_batch(self,texts):
    return np.array([self.encode(text) for text in texts])

  def getDict(self):
    voc = self.vectorizer.tokenizer.get_vocab()
    voc = dict(sorted(voc.items(), key=lambda item: item[1]))
    return voc
```

```python
def encode_df(dataset,tokenizer,vectorizer):
  train,val,test = dataset

  tokenizer.train(train)
  train_token = [tokenizer.encode(x) for x in train]
  val_token = [tokenizer.encode(x) for x in val]
  if test is not None:
    test_token = [tokenizer.encode(x) for x in test]

  vectorizer.train(train_token)
  train_vectors = vectorizer.encode_batch(train_token)
  val_vectors = vectorizer.encode_batch(val_token)
  if test is not None:
    test_vectors = vectorizer.encode_batch(test_token)
  else:
    test_vectors = None

  return train_vectors,val_vectors,test_vectors
```

```python
from tensorflow.keras.layers import Dense, Input, Dropout,concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import layers
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split

final_df = pd.concat([train_df,val_df,test_df])
final_train_df, final_val_df = train_test_split(final_df, test_size=0.1,random_state=69)
final_train_df_labels = final_train_df['Is Music'].values.astype(int)
final_val_df_labels = final_val_df['Is Music'].values.astype(int)

def compile_final():
  title_tokenizer = TokenizerNgram((1,3))
  title_vectorizer = VectorizerCount(8500)
  description_tokenizer = TokenizerWord()
  description_vectorizer = VectorizerTFIDF()
  category_tokenizer = TokenizerNone()
  category_vectorizer = VectorizerLabel()

  train_title_vectors, val_title_vectors, test_title_vectors = encode_df((final_train_df['Title'], final_val_df['Title'],None), title_tokenizer, title_vectorizer)
  train_desc_vectors, val_desc_vectors, test_desc_vectors = encode_df((final_train_df['Description'], final_val_df['Description'],None), description_tokenizer, description_vectorizer)
  train_cat_vectors, val_cat_vectors, test_cat_vectors = encode_df((final_train_df['Categories'], final_val_df['Categories'],None), category_tokenizer, category_vectorizer)
  train_duration = final_train_df["Duration"].values.astype(int)
  val_duration = final_val_df["Duration"].values.astype(int)

  title_input = Input(shape=(train_title_vectors.shape[1],))
  title_x = Dense(4, activation='elu')(title_input)
  title_x = Dropout(0.1)(title_x)
  title_x = Dense(2, activation='sigmoid')(title_x)
  title_x = Dropout(0.2)(title_x)

  desc_input = Input(shape=(train_desc_vectors.shape[1],))
  desc_x = Dense(64, activation='relu')(desc_input)
  desc_x = Dropout(0.1)(desc_x)


  cat_input = Input(shape=(train_cat_vectors.shape[1],))
  cat_x = cat_input

  dur_input = Input(shape=(1,))
  dur_x = Dense(256, activation='sigmoid')(dur_input)
  dur_x = Dropout(0.2)(dur_x)
  dur_x = Dense(32, activation='tanh')(dur_x)
  dur_x = Dropout(0.3)(dur_x)

  combined = layers.concatenate([title_x, desc_x, cat_input,dur_x])
  x = combined

  output = layers.Dense(1, activation='sigmoid')(x)

  model = Model(inputs=[title_input, desc_input,cat_input,dur_input], outputs=output)

  optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
  model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC()]
  )
  anti_overfit = EarlyStopping(monitor='val_loss',patience=20,restore_best_weights=True,min_delta=0.005,mode="min")
  model.fit(
      [train_title_vectors, train_desc_vectors,train_cat_vectors,train_duration],
      final_train_df_labels,
      epochs=5000,
      validation_data=([val_title_vectors, val_desc_vectors,val_cat_vectors,val_duration], final_val_df_labels),
      callbacks=[anti_overfit],)

  title_pipeline = (title_tokenizer, title_vectorizer)
  description_pipeline = (description_tokenizer, description_vectorizer)
  category_pipeline = (category_tokenizer, category_vectorizer)

  return model,title_pipeline,description_pipeline,category_pipeline

model,title_pipeline,description_pipeline,category_pipeline = compile_final()
```

```python
def calculate_confidence(predicted, threshold):
    if predicted >= threshold:
        confidence = (predicted - threshold) / (1 - threshold)
    else:
        confidence = (threshold - predicted) / threshold

    return round(confidence * 100, 2)
```
